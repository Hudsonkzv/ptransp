```python
import pandas as pd
import seaborn as sn
```

```python
# Carrega os dados do arquivo CSV
df = pd.read_csv("https://raw.githubusercontent.com/Hudsonkzv/ptransp/main/aux_pnad_tratada.csv")

# Formata a coluna de referência para o formato MMM-AA
df['mes_ref'] = pd.to_datetime(df['mes_ref'], format='%Y%m').dt.strftime('%b/%y')

# Adiciona uma coluna "valor_total_beneficio" para a soma dos valores das três categorias de benefício
df["valor_total_beneficio"] = df["valor_bolsa_familia"] + df["valor_cadunico"] + df["valor_extracad"]

# Filtra as colunas que serão avaliadas na correlação
df_filtrado = df[["mes_ref", "uf", "forca_conta_propria", "forca_trabalho_descoupada", "valor_total_beneficio", "valor_bolsa_familia", "valor_cadunico", "valor_extracad"]]
```

```python
# Análise de correlação

correlation = df_filtrado.corr()
```

```python
# Plot dos gráficos de dispersão

sn.pairplot(df_filtrado, diag_kws={'edgecolor':'k'}, plot_kws={'alpha':0.5, 'edgecolor':'k'})
```

```python
# Plot da matriz de correlação

plot = sn.heatmap(correlation, annot = True, fmt=".2f", linewidths=.5)
plot
```
